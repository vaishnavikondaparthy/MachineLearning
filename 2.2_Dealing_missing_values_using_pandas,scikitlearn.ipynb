```python
#dealing with missing values
```

```python
#replace "?" to NaN
df.replace("?",np.nan, inplace = True)
df.head(5)
```

```python
missing_data=df.isnull()
missing_data.head(5)
```

```python
for column in headers_data:
  print(column)
  print(missing_data[column].value_counts())
  print("")
```

```python
avg_norm_loss=df["normalized-losses"].astype('float').mean(axis=0)
print("Average of normalized-losses:",avg_norm_loss)
```

```python
df["normalized-losses"].replace(np.nan, avg_norm_loss,inplace = True)
```

```python
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Average of bore:", avg_bore)
```

```python
df["bore"].replace(np.nan,avg_bore,inplace=True)
```

```python
avg_stroke=df['stroke'].astype('float').mean(axis=0)
print("Average of stroke:", avg_stroke)
```

```python
df["stroke"].replace(np.nan,avg_stroke,inplace=True)
```

```python
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average of peak rpm:", avg_peakrpm)
```

```python
df["peak-rpm"].replace(np.nan,avg_peakrpm,inplace=True)
```

```python
avg_horsepower=df['horsepower'].astype('float').mean(axis=0)
print("Average of horsepower:", avg_horsepower)
```

```python
df["horsepower"].replace(np.nan,avg_horsepower,inplace=True)
```

```python
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].value_counts().idxmax()
df.head()
```

```python
df['num-of-doors'].replace(np.nan,"four",inplace=True)
df.head()
```

```python
# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"],axis=0,inplace=True)

#reset index,because we dropped two rows
df.reset_index(drop=True,inplace=True)
```

```python
missing_data=df.isnull()
missing_data.head(5)

for column in headers_data:
  print(column)
  print(missing_data[column].value_counts())
  print("")
```

```python
#dealing missing values using scikit-learn
```

```python
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```

```python
from sklearn.impute import SimpleImputer#same even for classification
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(df[['normalized-losses','stroke','bore','horsepower','peak-rpm']])
df[['normalized-losses','stroke','bore','horsepower','peak-rpm']] = imputer.transform(df[['normalized-losses','stroke','bore','horsepower','peak-rpm']])
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imputer.fit(df[['num-of-doors']])
df[['num-of-doors']] = imputer.transform(df[['num-of-doors']])
```

```python
# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
```

```python
missing_data = df.isnull()
missing_data.head(5)

for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")
```
